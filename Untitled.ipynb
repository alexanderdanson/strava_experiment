```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

steve_activities_file_name = os.path.join("activities_steve.csv")
alex_activities_file_name = os.path.join("activities.csv")

plt.style.use("classic")
plt.figure(figsize=(12.5, 4.25))

# Data import
alex_activities = pd.read_csv(alex_activities_file_name, index_col=0)
steve_activities = pd.read_csv(steve_activities_file_name, index_col=0)

#Adding a User column
alex_activities['User'] = "Alex"
steve_activities['User'] = "Steve"

# Changing column names and converting dates (different formats in the various CSVs)
alex_activities.rename(columns={'Activity Date': 'activity_date', 'Activity Type': 'activity_type', 'Elapsed Time': 'elapsed_time'}, inplace=True)
steve_activities.rename(columns={'Activity Date': 'activity_date', 'Activity Type': 'activity_type', 'Elapsed Time': 'elapsed_time'}, inplace=True)
steve_activities['activity_date'] = pd.to_datetime(steve_activities['activity_date'], format="%d %b %Y, %H:%M:%S")
alex_activities['activity_date'] = pd.to_datetime(alex_activities['activity_date'], format="%b %d, %Y, %I:%M:%S %p")

# Joining tables
activities = alex_activities.append(steve_activities)

#Convert distance to numbers
activities["Distance"] = pd.to_numeric(activities["Distance"], errors="coerce")

def create_bar_chart(dataset, y_column, title):
    grouped_activities = dataset.groupby("activity_type")
    totals = grouped_activities.sum()
    totals['{}'.format(y_column)].plot(kind="bar")
    plt.title("{}".format(title), fontsize=12)
    plt.xlabel("Activity Type", fontsize=10)
    plt.ylabel("{}".format(y_column), fontsize=10)
    #plt.savefig("{}.pdf".format(title))
    plt.show()
```

```python
create_bar_chart(activities.loc[activities['User'] == "Steve"], 'Distance', 'Total Distances')
create_bar_chart(activities.loc[activities['User'] == "Alex"], 'Distance', 'Total Distances')
```
